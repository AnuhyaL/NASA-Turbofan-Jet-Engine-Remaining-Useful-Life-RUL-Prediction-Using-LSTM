```python
from data_processing import load_cmapss, add_rul, create_sliding_windows
from models import RULPredictor
from utils import compute_metrics
import numpy as np
```

```python
df = load_cmapss("/content/train_FD003.txt")
df = add_rul(df, cap=125)
df.head()
```
